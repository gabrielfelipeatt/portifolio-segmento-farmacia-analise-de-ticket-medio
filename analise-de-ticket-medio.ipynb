```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Carregar os dados
df = pd.read_excel('/kaggle/input/dataset/dados.xlsx')
```

```python
# 2. Converter coluna de data para datetime
df['data_venda'] = pd.to_datetime(df['data_venda'])
```

```python
# 3. Criar coluna 'Ano-MÃªs'
df['Ano-MÃªs'] = df['data_venda'].dt.to_period('M')
```

```python
# 4. Calcular Ticket MÃ©dio: Total LÃ­quido / Quantidade
# (por mÃªs, para acompanhar a evoluÃ§Ã£o mensal)
ticket_medio_mensal = df.groupby('Ano-MÃªs').agg({
    'valor total liquido': 'sum',
    'quantidade': 'sum'
})
ticket_medio_mensal['Ticket MÃ©dio'] = ticket_medio_mensal['valor total liquido'] / ticket_medio_mensal['quantidade']
```

```python
# 5. Plotar grÃ¡fico do Ticket MÃ©dio mensal
plt.figure(figsize=(10, 5))
ticket_medio_mensal['Ticket MÃ©dio'].plot(marker='o', color='purple')
plt.title('ðŸ“Š EvoluÃ§Ã£o do Ticket MÃ©dio Mensal')
plt.xlabel('Ano-MÃªs')
plt.ylabel('R$ por unidade')
plt.grid(True)
plt.tight_layout()
plt.show()
```
